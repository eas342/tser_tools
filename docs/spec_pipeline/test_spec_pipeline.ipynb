This tutorial describes the basics of how to use the `tshirt` spectroscopic pipeline. This can also be run to test that the code has installed correctly.

## Import Code and Run Extraction


Import the code and set up the example parameters. The example parameter set is available at 
https://github.com/eas342/tshirt/blob/master/tshirt/parameters/spec_params/example_spec_parameters.yaml <br>
This will extract spectra from the images in `../example_data/spec_time_series/`

````python
from pipeline import spec_pipeline
#spec = spec_pipeline.spec()
spec = spec_pipeline.spec(paramFile='parameters/spec_params/example_spec_parameters.yaml')
spec.do_extraction(useMultiprocessing=True)
%matplotlib inline
````

## Plotting Results

Now you can plot an example spectrom from the time series.

````python
spec.plot_one_spec()
````

`plot_wavebin_series` will plot the time series for wavelength-binned data

````python
spec.plot_wavebin_series(savePlot=False)
````
